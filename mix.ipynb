```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = cv2.imread('./res/20230508-174531.png',cv2.IMREAD_UNCHANGED)
b = cv2.imread('./20230508-174535.png')
# f = cv2.cvtColor(f,cv2.COLOR_BGR2RGB)
# b = cv2.cvtColor(b,cv2.COLOR_BGR2RGB)
```

```python
img = f[...,:3]
```

```python
f_m = f[...,-1]
b_m = 255 - f_m
```

```python
f_m = f_m.astype(np.float32) / 255
b_m = b_m.astype(np.float32) / 255
```

```python
x = (f_m[...,None] * img + b_m[...,None] *b).astype(np.uint8)
x = cv2.cvtColor(x,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,12))
plt.axis('off')
plt.imshow(x)
```
